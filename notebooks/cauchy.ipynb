```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import cauchy

fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = cauchy.stats(moments='mvsk')
scale = 0.01
x = np.linspace(cauchy.ppf(0.01, scale=scale), cauchy.ppf(0.99, scale=scale), 100)
ax.set_ylim((0.0, 0.5))
ax.plot(x, cauchy.pdf(x, scale=scale), "r-", lw=5, alpha=0.6, label="cauchy pdf")
```
